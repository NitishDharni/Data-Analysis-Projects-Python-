```python
Alone=[]
for i in range(0,891):
    y = df.loc[i,'SibSp'] + df.loc[i,'Parch']
    if y == 0:
        Alone.append(0)
        i=i+1
    else:
        Alone.append(1)
        i=i+1
        
df['Num_FAM'] = Alone
df.head()
```
