```python
df_closing = web.DataReader(stock_list,'yahoo',start,end)['Adj Close']
df_closing.head()
```

```python
df_ret = df_closing.pct_change()
df_ret.head()
```

```python
import scipy.stats as stats
plt.figure(figsize=(15,4))
A = sns.jointplot('AMZN','GOOG',df_ret,kind='scatter',color='green', height=10)
A.annotate(stats.pearsonr)
```

```python
sns.pairplot(df_ret.dropna())
```

```python
RET = sns.PairGrid(df_ret.dropna())
RET.map_upper(sns.kdeplot,color='red')
RET.map_lower(plt.scatter,color='green')
RET.map_diag(plt.hist)
```

```python
CLS = sns.PairGrid(df_closing)
CLS.map_upper(sns.kdeplot,color='red')
CLS.map_lower(plt.scatter,color='green')
CLS.map_diag(plt.hist)
```
